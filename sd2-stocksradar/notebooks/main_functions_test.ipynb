## Add

```python
available = 'FB.BK'
```

```python
if available != None:
    table_name_suffix = ['_hourly_price_history','_daily_price_history','_weekly_price_history','_monthly_price_history']
    try:
        hour,day,week,month = (db_utils.read_table(available+table_name_suffix[0]),
                               db_utils.read_table(available+table_name_suffix[1]),
                               db_utils.read_table(available+table_name_suffix[2]),
                               db_utils.read_table(available+table_name_suffix[3]))
        if 'not found' not in hour and 'not found' not in day and 'not found' not in week and 'not found' not in month:
            result = hour,day,week,month
        else:
            raise Exception
    except:
        dfh, hname = data.get_hourly_data(available)
        dfd, dname = data.get_daily_data(available)
        dfw, wname = data.get_weekly_data(available)
        dfm, mname = data.get_monthly_data(available)
        db_utils.store_table(dfh,hname)
        db_utils.store_table(dfd,dname)
        db_utils.store_table(dfw,wname)
        db_utils.store_table(dfm,mname)
        result = (db_utils.read_table(hname),db_utils.read_table(dname),db_utils.read_table(wname),db_utils.read_table(mname))
else:
    result = None
```
