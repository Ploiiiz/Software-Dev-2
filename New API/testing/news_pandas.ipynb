```python
balance_sheet, meta = fundamentals.get_balance_sheet_quarterly('META')
balance_sheet = balance_sheet.sort_values(by='fiscalDateEnding',ascending=True)
balance_sheet
```

```python
balance_sheet.columns
```

```python
import matplotlib.pyplot as plt
```

```python
df = balance_sheet
df['totalAssets'] = df['totalAssets'].astype(int)
plt.bar(df['fiscalDateEnding'], df['totalAssets']/10e9)
plt.xlabel('Fiscal Year End')
plt.ylabel('Total Assets (Billions USD)')
plt.title('Total Assets by Fiscal Year End')
plt.show()
```

```python
df['currentLongTermDebt'] = df['currentLongTermDebt'].astype(int)
plt.bar(df['fiscalDateEnding'], df['currentLongTermDebt']/10e9)
plt.xlabel('Fiscal Year End')
plt.ylabel('Current Long Term Debt (Billions USD)')
plt.title('Current Long Term Debt by Fiscal Year End')
plt.show()
```
