```python
layout = go.Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,42,61,1)',
)
fig = go.Figure(data=[go.Candlestick(x=data['timestamp'],
                open=data['open'],
                high=data['high'],
                low=data['low'],
                close=data['close'])],
                layout=layout)
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)
fig.update_layout(
    updatemenus=[
        dict(
            type = "buttons",
            direction = "left",
            buttons=[
            {'label': '1D',
             'method': 'update',
             'args': [{'x': [data['timestamp']],
                       'open': [data['open']],
                       'high': [data['high']],
                       'low': [data['low']],
                       'close': [data['close']]}]},
            {'label': '1W',
             'method': 'update',
             'args': [{'x': [data1['timestamp']],
                       'open': [data1['open']],
                       'high': [data1['high']],
                       'low': [data1['low']],
                       'close': [data1['close']]}]},
            {'label': '1M',
             'method': 'update',
             'args': [{'x': [data2['timestamp']],
                       'open': [data2['open']],
                       'high': [data2['high']],
                       'low': [data2['low']],
                       'close': [data2['close']]}]},           
        ],
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.04,
            xanchor="left",
            y=1.3,
            yanchor="top"
        ),
    ]
)
fig.show()
```
